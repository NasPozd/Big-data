```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from yellowbrick.cluster import KElbowVisualizer, SilhouetteVisualizer
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN

#    
data = pd.read_csv('heart.csv', usecols=['age', 'trestbps', 'chol'])

#  
scaler = StandardScaler()
normalized_data = scaler.fit_transform(data)

#     K-means
kmeans = KMeans(random_state=42)
kmeans.fit(normalized_data)

#        
elbow_visualizer = KElbowVisualizer(kmeans, k=(2,10))
elbow_visualizer.fit(normalized_data)
elbow_visualizer.show()

#        
silhouette_visualizer = SilhouetteVisualizer(kmeans)
silhouette_visualizer.fit(normalized_data)
silhouette_visualizer.show()

#   
optimal_clusters = elbow_visualizer.elbow_value_

#       K-means
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['age'], data['trestbps'], data['chol'], c=kmeans.labels_, cmap='viridis', s=100, alpha=0.6, edgecolors='w')
ax.set_xlabel('Age')
ax.set_ylabel('Trestbps')
ax.set_zlabel('Chol')

#  
legend_elements = scatter.legend_elements()[0]
legend_labels = [f'Cluster {label}' for label in np.unique(hierarchical_clustering.labels_)]
ax.legend(legend_elements, legend_labels, loc='upper right')

plt.title('KMeans Clustering', loc='center', pad=20, color='red', fontsize=20, fontweight='bold')
plt.show()

#      
hierarchical_clustering = AgglomerativeClustering(n_clusters=optimal_clusters, affinity='euclidean', linkage='ward')
hierarchical_clustering.fit(normalized_data)

#  
plt.figure(figsize=(10, 8))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.title('Hierarchical Clustering Dendrogram', loc='center', pad=20, color='red', fontsize=20, fontweight='bold')
plt.xticks(fontsize=5, rotation=90)
plt.yticks(fontsize=10)
plt.tight_layout()  
plt.show()

#        
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['age'], data['trestbps'], data['chol'], c=hierarchical_clustering.labels_, cmap='viridis', s=100, alpha=0.6, edgecolors='w')
ax.set_xlabel('Age')
ax.set_ylabel('Trestbps')
ax.set_zlabel('Chol')

#  
legend_elements = scatter.legend_elements()[0]
legend_labels = [f'Cluster {label}' for label in np.unique(hierarchical_clustering.labels_)]
ax.legend(legend_elements, legend_labels, loc='upper right')

plt.title('Hierarchical Clustering', loc='center', pad=20, color='red', fontsize=20, fontweight='bold')
plt.show()

#     DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan.fit(normalized_data)

#       DBSCAN
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['age'], data['trestbps'], data['chol'], c=dbscan.labels_, cmap='viridis', s=100, alpha=0.6, edgecolors='w')
ax.set_xlabel('Age')
ax.set_ylabel('Trestbps')
ax.set_zlabel('Chol')

#  
legend_elements = scatter.legend_elements()[0]
legend_labels = [f'Cluster {label}' for label in np.unique(hierarchical_clustering.labels_)]
ax.legend(legend_elements, legend_labels, loc='upper right')

plt.title('DBSCAN Clustering', loc='center', pad=20, color='red', fontsize=20, fontweight='bold')
plt.show()
```
